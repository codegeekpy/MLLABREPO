```python
from sklearn.datasets import load_iris
import numpy as np
iris_dataset = load_iris()
iris_dataset.keys()
iris_dataset.DESCR
```

```python
X = iris_dataset['data']
y = iris_dataset['target']
print(iris_dataset.feature_names)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.4,random_state = 20)

print("X_train shape: {}".format(X_train.shape))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lin_reg = LinearRegression()

lin_reg.fit(X_train,y_train)
```

```python
y_pred = lin_reg.predict(X_test)

print("The score of test {:.2f}".format(lin_reg.score(X_test,y_pred)))
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Score :",score)
print("Mean Squared Error: {:.2f}".format(mse))
```

```python
import  matplotlib.pyplot as plt

r = iris_dataset.data[:,0].reshape(-1,1)
s = iris_dataset.data[:,2].reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split

S_train,S_test,r_train,r_test = train_test_split(s,r,test_size = 0.3,random_state=30)

print("S_train:{}".format(S_train.shape))
print("\nS_test:{}".format(S_test.shape))
print("\nr_train:{}".format(r_train.shape))
print("\nr_test:{}".format(r_test.shape))
```

```python
lin_reg.fit(S_train,r_train)
s_pred = lin_reg.predict(S_test)
```

```python
plt.scatter(S_test,r_test)
plt.plot(S_test,s_pred,color='red')
plt.show()
```
